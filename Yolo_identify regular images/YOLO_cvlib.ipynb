```
#Read file with cv package
import cvlib as cv
from cvlib.object_detection import draw_bbox
im = cv2.imread('image_001.jpg')
print(im)
```

```
#Use the pretrained model to identify and draw anchor boxes
bbox, label, conf = cv.detect_common_objects(im)
output_image = draw_bbox(im, bbox, label, conf)
plt.imshow(output_image)
plt.show()
```

```
#Loading second image to see prediction
im2 = cv2.imread('image_002.jpg')
print(im2)
```

```
#Use the model to predict on second image
bbox, label, conf = cv.detect_common_objects(im2)
output_image2 = draw_bbox(im2, bbox, label, conf)
plt.imshow(output_image2)
plt.show()
```

```
#Loading third image to see prediction
im3 = cv2.imread('image_003.jpg')
print(im3)
```

```
#Use the model to predict on third image
bbox, label, conf = cv.detect_common_objects(im3)
output_image3 = draw_bbox(im3, bbox, label, conf)
plt.imshow(output_image3)
plt.show()
```

```
#Loading fourth image to see prediction
im4 = cv2.imread('image_004.jpg')
print(im4)
```

```
#Use the model to predict on fourth image
bbox, label, conf = cv.detect_common_objects(im4)
output_image4 = draw_bbox(im4, bbox, label, conf)
plt.imshow(output_image4)
plt.show()
```
